**DROPPING THE UNNECESSARY COLUMNS for CUSTOMER dataset**

```python
df_cust_new=df_customer.drop(['POST_SUBSCRIBED','SMS_SUBSCRIBED','EMAIL_SUBSCRIBED','PHONE_SUBSCRIBED','SOCIAL_SUBSCRIBED','ANONYMISE'],axis=1)
df_cust_new.info() #Hence we remove these columns
```

```python
## TO FETCH UNIQUE COUNTRIES FOR ONE-HOT ENCODING ###

np.unique(df_cust_new.COUNTRY) # 
```

```python
def one_hot_encode(df, column_name):
    one_hot = pd.get_dummies(df[column_name])
    df = df.drop(column_name, axis=1)
    df = df.join(one_hot)
    return df
```

```python
one_hot_encode(df_cust_new,'COUNTRY').head()
```

```python
## Dropping the City Column as it has several unique values for One Hot Encoding, and Dropping the other cols as the contact info and names dont impact the recommender.##

df_cust_new1=df_cust_new.drop(['FIRST_NAME','LAST_NAME','EMAIL','PHONE_NUMBER','POST_CODE','CITY'],axis=1)
df_cust_new1.info() #Hence we remove these columns
```

```python
df_cust_new1=df_cust_new1.rename(columns={"ID": "c_id"})
df_cust_new1.head()
```
